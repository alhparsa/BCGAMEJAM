```python
import numpy as np
from scipy.io.wavfile import read
import sounddevice as sd
import matplotlib.pyplot as plt
```

```python
fs = 41800
sd.default.samplerate = fs 
sd.default.channels = 1
```

```python
dur = 2.5
str_rec = sd.rec(int(dur*fs))[30:13000]
sd.wait()

str_fft = np.fft.rfft(str_rec)
prev = np.argmax(str_fft)
```

```python
temp = 0
dur = 0.08
for i in range(20):
    wrestling_rec = sd.rec(int(dur*fs))[30:13000]
    sd.wait()

    wrestling_fft = np.fft.rfft(wrestling_rec)
    if (np.argmax(wrestling_fft) > prev + 200): 
        print("up")
        temp+=1
    elif (np.argmax(wrestling_fft) < prev - 200):
        print("down")
    else:
        print("constant")
    prev = np.argmax(wrestling_fft)
    
print(temp)
```
